```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
X.shape, y.shape
```

```python
kmeans = KMeans(n_clusters=4, n_init=10,random_state=5)
```

```python
kmeans.fit(X, y)
```

```python
labels = kmeans.labels_
```

```python
centers  =  kmeans.cluster_centers_
```

```python
inertia = kmeans.inertia_
inertia
```

```python
plt.scatter(X[:,0],X[:,1],label = 'Data points', c=kmeans.labels_, cmap='viridis')
plt.scatter(centers[:,0],centers[:,1],label = 'centriods',c='red', s=200, alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
