```python
import matplotlib.pyplot as plt
a = (20,50,30,100,70,90,50)
labels = ('Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday')
explode=(0,0,0,0.2,0,0,0.1)
plt.pie(a,labels=labels,explode=explode,shadow='True')
```
